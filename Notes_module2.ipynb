## The `DataFrame`

```python
import pandas as pd
```

```python
record1 = pd.Series({'Name': 'Alice',
                        'Class': 'Physics',
                        'Score': 85})
record2 = pd.Series({'Name': 'Jack',
                        'Class': 'Chemistry',
                        'Score': 82})
record3 = pd.Series({'Name': 'Helen',
                        'Class': 'Biology',
                        'Score': 90})
df = pd.DataFrame([record1, record2, record3],
                  index=['school1', 'school2', 'school1'])

df.head()
```

```python
students = [{'Name': 'Alice',
              'Class': 'Physics',
              'Score': 85},
            {'Name': 'Jack',
             'Class': 'Chemistry',
             'Score': 82},
            {'Name': 'Helen',
             'Class': 'Biology',
             'Score': 90}]

df = pd.DataFrame(students, index = ['school1', 'school2', 'school1'])
df
```

```python
df
```

```python
# Drop has two interesting optional parameters. The first is called inplace, and if it's 
# set to true, the DataFrame will be updated in place, instead of a copy being returned. 
# The second parameter is the axes, which should be dropped. By default, this value is 0, 
# indicating the row axis. But you could change it to 1 if you want to drop a column.
copy_df = df.copy()
copy_df.drop('Class', inplace = True, axis = 1)
copy_df
```

```python
# There is a second way to drop a column, and that's directly through the use of the indexing 
# operator, using the del keyword. This way of dropping data, however, takes immediate effect 
# on the DataFrame and does not return a view.
del copy_df['Name']
copy_df
```

## `DataFrame` Example


```python
import pandas as pd
```

```python
df = pd.read_csv('datasets/presidents.csv')
df.head(10)
```

```python
df['First'] = df['President'].replace('[ ].*', '', regex = True)
df.head(10)
```

```python
del(df['First'])

def split_name(row):
    row['First name'] = row['President'].split()[0]
    row['Surname'] = row['President'].split()[-1]
    return row

df = df.apply(split_name, axis = 'columns')
df.head(10)
```

```python
del(df['First name'])
del(df['Surname'])
```

```python
pattern = "(^[\w]*)(?:.* )([\w]*$)"

df['President'].str.extract(pattern).head(10)
```

```python
pattern = "(?P<First_name>^[\w]*)(?:.* )(?P<Surname>[\w]*$)"
names = df['President'].str.extract(pattern).head(10)
names
```

```python
df['First'] = names['First_name']
df['Second'] = names['Surname']
df.head(10)
```

```python
df['Born'] = df['Born'].str.extract('([\w]{3} [\w]{1,2}, [\w]{4})')
df['Born'].head(10)
```

```python
pd.to_datetime(df['Born']).head(10)
```
