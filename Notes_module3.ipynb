## Merging

```python
import pandas as pd
```

```python
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR'},
                         {'Name': 'Sally', 'Role': 'Course liasion'},
                         {'Name': 'James', 'Role': 'Grader'}])
student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally', 'School': 'Engineering'}])
staff_df = staff_df.set_index('Name')
student_df = student_df.set_index('Name')
print(staff_df.head())
print(student_df.head())
```

```python
pd.merge(staff_df, student_df, how = 'outer', right_index = True, left_index = True)
```

```python
pd.merge(staff_df, student_df, how = 'inner', right_index = True, left_index = True)
```

```python
pd.merge(staff_df, student_df, how = 'left', right_index = True, left_index = True)
```

```python
pd.merge(staff_df, student_df, how = 'right', right_index = True, left_index = True)
```

```python
staff_df = staff_df.reset_index()
student_df = student_df.reset_index()

pd.merge(staff_df, student_df, how = 'right', on = 'Name')
```

```python
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role': 'Director of HR', 
                          'Location': 'State Street'},
                         {'Name': 'Sally', 'Role': 'Course liasion', 
                          'Location': 'Washington Avenue'},
                         {'Name': 'James', 'Role': 'Grader', 
                          'Location': 'Washington Avenue'}])
student_df = pd.DataFrame([{'Name': 'James', 'School': 'Business', 
                            'Location': '1024 Billiard Avenue'},
                           {'Name': 'Mike', 'School': 'Law', 
                            'Location': 'Fraternity House #22'},
                           {'Name': 'Sally', 'School': 'Engineering', 
                            'Location': '512 Wilson Crescent'}])

pd.merge(staff_df, student_df, how = 'left', on = 'Name')
```

```python
import pandas as pd
import numpy as np
import timeit
```

```python
df = pd.read_csv('datasets/census.csv')
df.head(10)
```

```python
(df.where(df['SUMLEV'] == 50)
    .dropna()
    .rename(columns = {'STNAME': 'Name of state', 'CTYNAME': 'city'})
    .set_index(['Name of state', 'city']))
```

```python
df = df[df['SUMLEV'] == 50]
df.set_index(['STNAME', 'CTYNAME'], inplace = True)
df.rename(columns = {'ESTIMATESBASE2010': 'Estimates base of 2010'})
```

```python
def first_approach():
    global df
    return (df.where(df['SUMLEV'] == 50)
            .dropna()
            .rename(columns = {'STNAME': 'Name of state', 'CTYNAME': 'city'})
            .set_index(['Name of state', 'city']))

df = pd.read_csv('datasets/census.csv')
timeit.timeit(first_approach, number = 10)
```

```python
def second_approach():
    global df
    new_df = df[df['SUMLEV'] == 50]
    new_df.set_index(['STNAME', 'CTYNAME'], inplace = True)
    return new_df.rename(columns = {'ESTIMATESBASE2010': 'Estimates base of 2010'})

df = pd.read_csv('datasets/census.csv')
timeit.timeit(second_approach, number = 10)
```

```python
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min': np.min(data), 'max': np.max(data)})

df.apply(min_max, axis = 'columns').head()
```

```python
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['max'] = np.max(data)
    row['min'] = np.min(data)
    return row
df.apply(min_max, axis = 'columns').head()
```

```python
def get_state_region(x):
    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 
                 'Rhode Island','Vermont','New York','New Jersey','Pennsylvania']
    midwest = ['Illinois','Indiana','Michigan','Ohio','Wisconsin','Iowa',
               'Kansas','Minnesota','Missouri','Nebraska','North Dakota',
               'South Dakota']
    south = ['Delaware','Florida','Georgia','Maryland','North Carolina',
             'South Carolina','Virginia','District of Columbia','West Virginia',
             'Alabama','Kentucky','Mississippi','Tennessee','Arkansas',
             'Louisiana','Oklahoma','Texas']
    west = ['Arizona','Colorado','Idaho','Montana','Nevada','New Mexico','Utah',
            'Wyoming','Alaska','California','Hawaii','Oregon','Washington']
    
    if x in northeast:
        return "Northeast"
    elif x in midwest:
        return "Midwest"
    elif x in south:
        return "South"
    else:
        return "West"
df['state_region'] = df['STNAME'].apply(lambda x: get_state_region(x))
df[['STNAME','state_region']]
```

## Grouping Data

### Splitting

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/census.csv')
df = df[df['SUMLEV']==50]
df.head()
```

```python
%%timeit -n 3
for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME'] == state).dropna()['CENSUS2010POP'])
    print('Counties in state ' + state + 
          ' have an average population of ' + str(avg))
```

```python
%%timeit -n 3
for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    print('Counties in state ' + group + 
          ' have an average population of ' + str(avg))
```

```python
df.groupby('STNAME')['CENSUS2010POP'].mean() 
```

```python
df = df.set_index('STNAME')

def set_batch_number(item):
    if item[0]<'M':
        return 0
    if item[0]<'Q':
        return 1
    return 2

for group, frame in df.groupby(set_batch_number):
    print('There are ' + str(len(frame)) + ' records in group ' + str(group) + ' for processing.')
```

```python
df = pd.read_csv('datasets/listings.csv')
df.head()
```

```python
df = df.set_index(['cancellation_policy', 'review_scores_value'])
for group, frame in df.groupby(level = (0,1)):
    policy, score = group
    print((policy, float(score)))
```

```python
def grouping_fun(item):
    if item[1] == 10.0:
        return (item[0], 'is the best')
    else:
        return (item[0], "isn't 10.0")
for group, frame in df.groupby(by = grouping_fun):
    print(group)
```

```python
df.head()
```

```python
cols=['cancellation_policy','review_scores_value']
transform_df = df[cols].groupby('cancellation_policy').transform(np.nanmean)
transform_df.head()
```

```python
transform_df.rename({'review_scores_value':'mean_review_scores'}, axis='columns', inplace=True)
df=df.merge(transform_df, left_index=True, right_index=True)
print(transform_df.head())
df.head()
```

```python
df['mean_diff'] = np.absolute(df['review_scores_value']-df['mean_review_scores'])
df['mean_diff'].head()
```

### Filtering

```python
df.groupby('cancellation_policy').filter(lambda x: np.nanmean(x['review_scores_value']) > 9.2)
```

## Pivot Table

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/cwurData.csv')
df.head()
```

```python
def add_col(ranking):
    if 1 <= ranking <= 100:
        return "First Tier Top University"
    elif 101 <= ranking <= 200:
        return "Second Tier Top University"
    elif 201 <= ranking <= 300:
        return "Third Tier Top University"
    else:
        return "Other Top University"

df['Rank_Level'] = df['world_rank'].apply({lambda x: add_col(x)})
df
```

```python
bins = [0, 100, 200, 300, float('inf')]
labels = ["First Tier Top University",
          "Second Tier Top University",
          "Third Tier Top University",
          "Other Top University"]
df['Rank_Level'] = pd.cut(df['world_rank'], bins=bins, labels=labels)
df
```

```python
df.pivot_table(values = 'score', index = 'country', columns = 'Rank_Level', aggfunc = [np.mean])
```

```python
df.pivot_table(values = 'score', index = 'country', columns = 'Rank_Level', aggfunc = [np.mean, np.max])
```

```python
df.pivot_table(values = 'score', index = 'country', columns = 'Rank_Level', aggfunc = [np.mean, np.max], 
              margins = True)
```

```python
new_df=df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.mean, np.max], 
               margins=True)
print(new_df.index)
print(new_df.columns)
```
